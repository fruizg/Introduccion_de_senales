## Lectura y gráfica de una Señal

### Importación de librerias

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import re
from scipy import signal
```

##### Abrir archivo txt

```python
f = open("SenalECG/ECG_reposo_sujeto2.txt","r")
raw_data = f.readlines()[1] # con f.read() leemos todo el contenido
f.close()

raw_data
```

```python
x = re.findall("[0-5][0-9][0-9]\d", raw_data)

print(x)
```

```python
Fs = float(x[0])
Ts=1/Fs

print(f" Fs={Fs} hz\n Ts={Ts} s")
```

## Leer archivo excluyendo 2 primeras filas

```python
SenalReposo = np.genfromtxt("SenalECG/ECG_reposo_sujeto2.txt", delimiter="\t",skip_header = 3)
SenalEjercicio = np.genfromtxt("SenalECG/ECG_ejercicio_S2.txt", delimiter="\t", skip_header=3)
```

## Visualización de la señal

```python
#Señal en reposo -Normalizacion
y1 = (SenalReposo[:, 5])
y2 = (SenalEjercicio[:, 5])
y1 = y1  - np.min(y1)
y1 = y1 / np.max(y1)
t1 = (np.arange(0,len(y1)) / Fs)
ind_peaks1, _ = signal.find_peaks(y1, height= 0.8)
ind_time1 = t1[ind_peaks1]

#Señal Después de actividad física - Normalizacion
y2 = y2 - np.min(y2)
y2 = y2 / np.max(y2)
y2 = -y2
t2 = (np.arange(0,len(y2)) / Fs)
ind_troughs, _ = signal.find_peaks(y2, height= -0.25)
ind_time2 = t2[ind_troughs]
y2 = -y2

#Graficos señales
plt.figure()
plt.subplot(2,1,1)
plt.subplots_adjust(hspace=0.7)
plt.plot(y1, label="señal Reposo") 
plt.plot(ind_peaks1, y1[ind_peaks1], 'ro')
plt.title("Señal ECG Reposo - Sujeto 2")    # graficamos la señal
plt.grid(linestyle=":")
plt.xlabel("Tiempo (ms)")
plt.xlim(0,4000)

plt.subplot(2,1,2)
plt.plot(y2, label="señal Actividad Físida")  
plt.plot(ind_troughs, y2[ind_troughs], 'go')
plt.title("Señal ECG Después de actividad física - Sujeto 2")    # graficamos la señal
plt.grid(linestyle=":")
plt.xlabel("Tiempo (ms)")
plt.xlim(0,4000)

fig = plt.gcf()
fig.set_facecolor('white')
#plt.savefig('ECG_Sujeto2_comparison.png', transparent=False, )
plt.show()
```

#### Hallando la frecuencia cardíaca

```python
fc1 = int(60 / (np.diff(ind_peaks1[0:2]) * (1/Fs)))
fc2 = int(60 / (np.diff(ind_troughs[0:2]) * (1/Fs)))
print (f"La frecuencia cardiaca del sujeto 1 en reposo fue: {fc1}; mientras que su frecuencia cardiaca después del ejercicio fue {fc2}")
```

## Análisis del dominio de la frecuencia

```python
X1 = np.fft.fft(y1) #Transformada de fourier a la primera señal - ECG en reposo
X1 = np.abs(X1)
X1 = 20 * np.log10(X1)
N1 = len(y1)
f1 = (Fs)*(np.arange(1,N1+1)/N1)

X2 = np.fft.fft(y2) #Transformada de Fourier a la segunda señal - ECG después de actividad física
X2 = np.abs(X2)
X2 = 20 * np.log10(X2)
N2 = len(y2)
f2 = (Fs)*(np.arange(1,N2+1)/N2)
#Graficamos las transformadas para ver su comportamiento
plt.figure()
plt.subplot(2,1,1)
plt.subplots_adjust(hspace=0.7)
plt.title("FFT señal ECG en reposo")    # graficamos la señal
plt.grid(linestyle=":")
plt.xlabel("Frecuencia (Hz)") 
plt.ylabel("decibeles (db)")
plt.plot(f1,X1)
plt.xlim((0, Fs/2))


plt.subplot(2,1,2)
plt.title("FFT señal ECG después de actividad física")    # graficamos la señal
plt.grid(linestyle=":")
plt.xlabel("Frecuencia (Hz)") 
plt.ylabel("decibeles (db)")
plt.plot(f2,X2)
plt.xlim((0, Fs/2))
fig = plt.gcf()
fig.set_facecolor('white')
#plt.savefig('FFT_Sujeto2.png', transparent=False, )
```
